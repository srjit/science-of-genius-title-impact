```python
import matplotlib.pyplot as plt
from scipy.stats import gamma
import numpy as np
```

```python
a =3.5

r = gamma.rvs(a, size=1000)
alpha, loc, beta = gamma.fit(r)


x = np.linspace(0,10,10)
y1 = gamma.pdf(x, a=alpha)

f = plt.figure(figsize=(20,15))
ax = f.add_subplot(331)
ax.scatter(x, y1)
ax.set_xlabel("Variation")
ax.set_ylabel("Probability")
#ax.plot(x, y1, "k-")
```

```python
fig, ax = plt.subplots(1, 1)
x = np.linspace(gamma.ppf(0.01, a),
                gamma.ppf(0.99, a), 100)

ax.plot(x, gamma.pdf(x, a),
       'r-', lw=5, alpha=0.9, label='gamma pdf')
# ax.hist(r, density=True, histtype='stepfilled', alpha=0.2)
# ax.legend(loc='best', frameon=False)
# plt.show()
```

```python
import scipy.stats as stats    

alpha = 5
loc = 100.5
beta = 22
data = stats.gamma.rvs(alpha, loc=loc, scale=beta, size=10000)  
```

```python
from collections import Counter
counts = Counter(list(map(lambda x: round(x, 0), data))).items()


X = [x for (x, y) in counts]
Y = [y for (x, y) in counts]
Y2 = [y/sum(Y) for y in Y]
```
